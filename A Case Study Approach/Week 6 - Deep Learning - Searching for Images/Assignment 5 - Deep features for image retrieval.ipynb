```python
dog_test = image_test.filter_by("dog","label")
cat_test = image_test.filter_by("cat","label")
bird_test = image_test.filter_by("bird","label")
automobile_test = image_test.filter_by("automobile","label")

dog_test_model = gl.nearest_neighbors.create(dog_test,features=["deep_features"],label="id")
cat_test_model = gl.nearest_neighbors.create(cat_test,features=["deep_features"],label="id")
bird_test_model = gl.nearest_neighbors.create(bird_test,features=["deep_features"],label="id")
automobile_test_model = gl.nearest_neighbors.create(automobile_test,features=["deep_features"],label="id")

dog_cat_neighbors = cat_test_model.query(dog_test, k=1)
cat_bird_neighbors = bird_test_model.query(cat_test, k=1)
bird_automobile_neighbors = automobile_test_model.query(bird_test, k=1)
automobile_dog_neighbors = dog_test_model.query(automobile_test, k=1)
automobile_cat_neighbors = cat_test_model.query(automobile_test, k=1)
dog_bird_neighbors = bird_test_model.query(dog_test, k=1)

dog_dog_neighbors = dog_test_model.query(dog_test, k=1)
cat_cat_neighbors = cat_test_model.query(cat_test, k=1)
bird_bird_neighbors = bird_test_model.query(bird_test, k=1)
automobile_automobile_neighbors = automobile_test_model.query(automobile_test, k=1)
```

```python
dog_cat_neighbors.head()
```

```python
dog_distances = gl.SFrame({"dog-cat": dog_cat_neighbors["distance"],
                          "dog-automobile":automobile_dog_neighbors["distance"],
                          "dog-bird":dog_bird_neighbors["distance"],
                          "dog-dog": dog_dog_neighbors["distance"]})
```

```python
dog_distances.head()
```

```python
def is_dog_correct(row):
    if row["dog-automobile"] > row["dog-dog"] and row["dog-bird"] > row["dog-dog"] and row["dog-cat"] > row["dog-dog"]:
        return 1
    else:
        return 0
```

```python
dog_distances.apply(is_dog_correct).sum()
```
